```python
#intro


'''
Name: Aarav Chauhan
Roll: 2501420027
Date: 07 OCT 2025
Project name: Daily Calories Tracker CLI
'''


print('Welcome to the daily calorie tracker!\n')
print('This tool will help you log your meals, count calories, and check against your daily limit.\n\n')


#inputs and data collection

meal_names = []
calorie_amount = []


#to avoid invalid inputs


while True:
    x = input('How many meals did you eat today:\t')
    try:
        num_meals = int(x)
        if num_meals < 0:
            print('\nInvalid input. Please enter a positive number.\n')
        else:
            break
    except ValueError:
        print('\nInvalid input. Please enter a number.\n')


#to take inputs and consecutively add them to the list


for i in range(num_meals):
    meal = input(f"\n\nEnter meal {i+1} name:\t")
    meal_names.append(meal)
    calories = int(input(f"\nEnter calories for {meal}:\t"))
    calorie_amount.append(calories)



    #calorie calculations


total_calories = sum(calorie_amount)


#To check if number of meals is greater than zero before calculating average


if int(num_meals) > 0:
    average_calories = total_calories / int(num_meals)
    print(f"\n\nTotal calories consumed: {total_calories}\n")
    print(f"Average calories per meal: {average_calories}")
else:
    print("\n\nNo meals were entered, so average calories cannot be calculated.")


#Taking calorie limit input and comparing


limit = float(input('\n\nEnter your daily calorie limit:\t'))

under_limit = total_calories <= limit



#exceed limit waning system


if total_calories > limit:
  print("\nWARNING!!! YOU HAVE EXCEEDED THE DAILY CALORIE LIMIT")
else:
  print("\nGood Job! Your intake is within the daily calorie limit")



print('\n\n\nMeal Name\t\t\tCalories\n')
print('----------------------------------------------\n')
for i in range(num_meals):
    print(f"  {meal_names[i]}\t\t\t         {calorie_amount[i]}\n")
print(' \nTotal\t\t\t\t',total_calories)
print(' \nAverage\t\t\t\t',average_calories)
```
